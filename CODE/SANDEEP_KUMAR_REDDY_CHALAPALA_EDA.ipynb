```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
Transactions_data=pd.read_csv('gdrive/My Drive/Zeotap/Transactions.csv')
Transactions_data.head()
```

```python
Transactions_data.info()
```

```python
Transactions_data.describe()
```

```python
Transactions_data.isnull().sum()
```

```python
# Check for duplicate rows
print("\nDuplicate Rows:")
print(Transactions_data.duplicated().sum())
```

```python
# Explore categorical features for Transactions_data
categorical_cols = Transactions_data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nUnique values in {col}:")
    print(Transactions_data[col].unique())
    print(f"\nValue counts in {col}:")
    print(Transactions_data[col].value_counts())
```

```python
import pandas as pd
import numpy as np

numerical_cols = Transactions_data.select_dtypes(include=np.number).columns
num_numerical_cols = len(numerical_cols)

print(f"\nThe number of numerical columns in Transactions_data is: {num_numerical_cols}")
```

```python
if not numerical_cols.empty:
    for col in numerical_cols:
        print(f"\nStatistics for {col}:")
        print(Transactions_data[col].describe())
else:
    print("No numerical data found in the DataFrame.")
```

```python
if not numerical_cols.empty:
    for col in numerical_cols:
        print(f"\nStatistics for {col}:")
        print(Transactions_data[col].describe())
        # Plot the distribution of numerical features
        plt.figure(figsize=(8, 6))
        sns.histplot(Transactions_data[col], kde=True)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
else:
    print("No numerical data found in the DataFrame.")
```

```python
# Correlation matrix for numerical features
correlation_matrix = Transactions_data[numerical_cols].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
if len(numerical_cols) > 1:
    plt.figure(figsize=(10, 8))
    sns.heatmap(Transactions_data[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix of Numerical Features for Transactions')
    plt.show()
elif len(numerical_cols) == 1:
    print("Only one numerical column found, correlation analysis not applicable.")
else:
    print("No numerical columns found for correlation analysis.")
```

```python
# Convert 'TransactionDate' to datetime objects
Transactions_data['TransactionDate'] = pd.to_datetime(Transactions_data['TransactionDate'])

# Analyze transaction trends over time
# Group transactions by month and count the number of transactions
monthly_transactions = Transactions_data.groupby(Transactions_data['TransactionDate'].dt.to_period('M'))['TransactionID'].count()
plt.figure(figsize=(12,6))
plt.plot(monthly_transactions.index.to_timestamp(), monthly_transactions.values)
plt.title('Monthly Transaction Volume')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Explore relationship between 'Quantity' and 'Price'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Quantity', y='Price', data=Transactions_data)
plt.title('Quantity vs. Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()
```

```python
# Analyze the distribution of transaction amounts
plt.figure(figsize=(10, 6))
sns.histplot(Transactions_data['Price'], kde=True, bins=30)  # Increased bins for better granularity
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Explore the relationship between transaction amount and quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Price', data=Transactions_data, hue='ProductID') #Added hue to see product ID's
plt.title('Transaction Amount vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Transaction Amount')
plt.show()
```

```python
#Analyze transaction frequency per customer
customer_transaction_counts = Transactions_data['CustomerID'].value_counts()
plt.figure(figsize=(10,6))
sns.histplot(customer_transaction_counts, bins=30, kde=True)
plt.title('Distribution of Transactions per Customer')
plt.xlabel('Number of Transactions')
plt.ylabel('Number of Customers')
plt.show()
```

```python
#Further analysis on transaction date
# Daily transaction trends
daily_transactions = Transactions_data.groupby(Transactions_data['TransactionDate'].dt.date)['TransactionID'].count()
plt.figure(figsize=(12, 6))
plt.plot(daily_transactions.index, daily_transactions.values)
plt.title('Daily Transaction Volume')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Analyze transaction data by product
product_transactions = Transactions_data.groupby('ProductID')['TransactionID'].count().sort_values(ascending = False)
print(product_transactions.head(10)) #top 10 product transactions
plt.figure(figsize=(10,6))
product_transactions.head(10).plot(kind='bar') #Plot top 10 products
plt.title('Top 10 Products by Transactions')
plt.xlabel('Product ID')
plt.ylabel('Number of Transactions')
plt.show()
```
